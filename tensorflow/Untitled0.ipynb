```python
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import scipy.sparse

# Extract features (X) and target (y)
activities = mock_data['activities']
names_data = [activity['name'] for activity in activities] # New: extract names
domeinen_data = [activity['domeinen'] for activity in activities]
layers_data = [activity['layers'] for activity in activities]

# Encode 'domeinen' using MultiLabelBinarizer
mlb_domeinen = MultiLabelBinarizer()
X_domeinen_encoded = mlb_domeinen.fit_transform(domeinen_data)

# Encode 'name' using TfidfVectorizer
# Using a small vocabulary size to keep the feature space manageable for this small dataset
tfidf_vectorizer = TfidfVectorizer(max_features=10) # Limit features to avoid sparsity issues with very small dataset
X_names_encoded = tfidf_vectorizer.fit_transform(names_data)

# Convert TF-IDF output to a dense array if it's not already
if isinstance(X_names_encoded, scipy.sparse.csr.csr_matrix):
    X_names_encoded = X_names_encoded.toarray()

# Concatenate the encoded 'name' features with 'domeinen' features
X_encoded = np.hstack((X_domeinen_encoded, X_names_encoded))

# Create MultiLabelBinarizer for layers (target)
mlb_layers = MultiLabelBinarizer()
y_encoded = mlb_layers.fit_transform(layers_data)

print("Shape of encoded features (X):", X_encoded.shape)
print("Shape of encoded target (y):", y_encoded.shape)

print("\nExample of encoded X (first 2 rows):\n", X_encoded[:2])
print("\nExample of encoded y (first 2 rows):\n", y_encoded[:2])

print("\nDomeinen classes:", mlb_domeinen.classes_)
print("\nName TF-IDF features (first 5):", tfidf_vectorizer.get_feature_names_out()[:5])
print("\nLayers classes:", mlb_layers.classes_)
```

```python
# Example new input
new_activity_name = "new incident in hospital"
new_activity_domeinen = ["gezondheidzorg", "brandweer"]

# Preprocess new 'name' input using the same TF-IDF vectorizer
new_name_encoded = tfidf_vectorizer.transform([new_activity_name])

# Preprocess new 'domeinen' input using the same MultiLabelBinarizer
# Note: mlb_domeinen.transform expects a list of lists
new_domeinen_encoded = mlb_domeinen.transform([new_activity_domeinen])

# Concatenate the preprocessed features
# Ensure it's a dense array for the model input
new_X_encoded = np.hstack((new_domeinen_encoded, new_name_encoded.toarray()))

print(f"Shape of new_X_encoded: {new_X_encoded.shape}")
print(f"Example of new_X_encoded (first row):\n{new_X_encoded[0]}")
```

```python
# Make a prediction with the scaled model
predictions = model_scaled.predict(new_X_encoded)

# The output are probabilities, we need to convert them to binary labels
# A common threshold is 0.5 for sigmoid activations
predicted_labels_binary = (predictions > 0.5).astype(int)

# Inverse transform the binary labels to get the original layer names
predicted_layers = mlb_layers.inverse_transform(predicted_labels_binary)

print(f"\nInput Activity Name: {new_activity_name}")
print(f"Input Activity Domeinen: {new_activity_domeinen}")
print(f"Predicted Layers: {predicted_layers[0]}")
```
